# Clean CNP Prediction Only (Use Existing Trained Model)

This notebook runs only prediction using a previously trained model checkpoint.
It does not run training.


```python
from pathlib import Path
import importlib
import pandas as pd
from IPython.display import Image, display

import cnp_clean_pipeline as cnp_clean_pipeline
cnp_clean_pipeline = importlib.reload(cnp_clean_pipeline)
load_runtime_config = cnp_clean_pipeline.load_runtime_config
predict_cnp = cnp_clean_pipeline.predict_cnp
```

```python
# Update if needed
CONFIG_PATH = Path('../xenon/settings2.yaml').resolve()
runtime = load_runtime_config(CONFIG_PATH, seed=42)

print('config      :', runtime.config_path)
print('predict_dirs:', runtime.predict_dirs)
print('out_dir     :', runtime.out_dir)
print('epochs      :', runtime.epochs)
```

## Load Existing Model
Set this to your trained checkpoint path.


```python
# Example: cnp_v101.0_model_15epochs.pth
MODEL_PATH = runtime.out_dir / f'cnp_{runtime.version}_model_{runtime.epochs}epochs.pth'
print('model_path:', MODEL_PATH)
print('exists    :', MODEL_PATH.exists())
```

## Predict
If you want speed closer to old notebook behavior, set `mc_samples=1`.


```python
predict_result = predict_cnp(
    runtime,
    model_path=MODEL_PATH,
    mc_samples=1,
    chunk_size=30000,
)
predict_result
```
